```python
features = ['id', 'acc_x', 'acc_y', 'acc_z', 'gy_x', 'gy_y', 'gy_z']
X_data = train_feature[features].groupby('id').agg(['sum', 'median', 'max', 'min', 'mean'])  #id별로 잘 합쳐짐
X_exam = test_feature[features].groupby('id').agg(['sum', 'median','max', 'min', 'mean'])    #id별로 잘 합쳐짐
```

```python
Y_data = train_label['label']
Y_data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_data, Y_data, test_size=0.2,
                                                    random_state=26, stratify = Y_data)
# stratify  = Y가 한쪽으로 쏠리지 않게 해줌. 이게 없으면 성능차이가 좀 있을 수 있음.
```

```python
#RandomForest
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
clf = RandomForestClassifier(random_state =26) #인스턴스
clf.fit(X_train, Y_train) # 학습
pred = clf.predict(X_test)
accuracy = accuracy_score(Y_test, pred)
print(accuracy)
```

```python
#GridSearch : 교차검증+최적의 파라미터찾기  # 정확도는 떨어지지만 일반화는 좋음
```

```python
from sklearn.model_selection import GridSearchCV 
```

```python
%%time 

from sklearn.model_selection import GridSearchCV

params = {
    'n_estimators':[50, 100, 150],  #생성트리 개수
    'criterion':['gini', 'entropy'], #지니계수, 
    'max_depth': [6, 8, 10, 12],
    'min_samples_leaf': [8, 12, 18],
    'min_samples_split': [8, 16, 20]
}

rf_clf = RandomForestClassifier(random_state=0, n_jobs=-1)
grid_cv = GridSearchCV(rf_clf, param_grid=params, cv=2, n_jobs=-1)
grid_cv.fit(X_train, Y_train)
```

```python
print('최적 하이퍼 파라미터: \n', grid_cv.best_params_)
```

```python
pred = grid_cv.best_estimator_.predict(X_test)
accuracy = accuracy_score(Y_test, pred)
print('accuracy : {:.4f}'.format(accuracy))  #GridSearchCV : 정확도는 떨어지지만 일반화는 좋음
                                            #학습데이터 떨어짐.
```

RNN : 이전의 정보를 복사해서 과거 정보 바탁 예측(시계열)

LSTM : RNN의 한 종류. '장기의존성 문제'를 보완. 먼 과거의 이벤트를 고려하지 못하는 문제

GRU : LSTM의 간소화된 버전

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, LSTM   
from keras.layers import Dense, GRU
```

```python
train_label = pd.read_csv('./data/train_labels.csv')
train_feature = pd.read_csv('./data/train_features.csv')
test_feature = pd.read_csv('./data/test_features.csv')
sample_submission = pd.read_csv('./data/sample_submission.csv')
```

```python
import numpy as np
```

```python
train_feature
```

```python
X = tf.reshape(np.array(train_feature.iloc[:,2:]),[-1,600,6]) #3차원을 배열한다
X.shape
```

```python
train_feature.iloc[:,2:]
```

```python
# utils.to_categorical :    
#'label 칼럼을 One_hot encoding하여 (3125, 61)으로 재배열: 운동의 종류가 61
# One hot encoding이란, o,x를 1로 바꿈. 이런걸로
Y=tf.keras.utils.to_categorical(train_label['label'])
Y.shape
```

```python
model =Sequential()
model.add(LSTM(32, input_shape = (600,6)))   #퍼셉트론 개수 32
model.add(Dense(128, activation = 'relu'))
model.add(Dense(61, activation = 'softmax'))   #출력개수 61개, 이중뷴류는 Sigmoid, 다중분류는 softmax
model.compile(optimizer = 'rmsprop', loss='categorical_crossentropy',
             metrics=['accuracy'])  #rmsprop하고 adam 차이 rmsprop은 0.508, adam(relu)은 0.5120
#loss는 손실함수. 훈련에 사용. y-y^, 실제 예측과 괴리 얼마나 되는지 비용함수
#crossentropy는 분류문제에 쓴다.
#metric은 평가지표. 훈련 과정을 모니터링하는데 사용
```

```python
model.fit(X,Y,epochs=30, batch_size=128, validation_split=0.2)
```

```python
test_X = tf.reshape(np.array(test_feature.iloc[:,2:]),[-1,600,6])
test_X.shape
```

```python
prediction=model.predict(test_X)
```
